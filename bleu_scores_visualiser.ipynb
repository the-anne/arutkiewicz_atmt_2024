```python
import matplotlib.pyplot as plt
import numpy as np

beam_sizes = [1, 2, 3, 4, 6, 8, 9, 10, 12, 14]

bleu_scores = [17.1, 19.0, 19.1, 18.6, 19.7, 20.8, 22.0, 22.2, 22.2, 22.6]
brevity_penalty = [1, 1, 1, 1, 1, 1, 1, 1, 1, 0.975]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Beam Size')
ax1.set_ylabel('BLEU Score', color=color)
ax1.plot(beam_sizes, bleu_scores, color=color)
ax1.tick_params(axis='y', labelcolor=color)

for x, y in zip(beam_sizes, bleu_scores):
    ax1.annotate(f'{y:.1f}', (x, y), textcoords="offset points", xytext=(-10, 5), ha='center', color=color)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Brevity Penalty', color=color)  # we already handled the x-label with ax1
ax2.plot(beam_sizes, brevity_penalty, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
